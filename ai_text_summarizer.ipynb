```python
from transformers import pipeline

# Load a small pre-trained summarization model
summarizer = pipeline(
    "summarization",
    model="sshleifer/distilbart-cnn-12-6",
    tokenizer="sshleifer/distilbart-cnn-12-6"
)
```

```python
def summarize_text(text, min_len=60, max_len=140):
    if len(text.split()) < 50:
        return "Please enter at least 50 words."

    # Run the summarizer
    summary = summarizer(text, min_length=min_len, max_length=max_len, do_sample=False)
    return summary[0]['summary_text']
```

```python
sample_text = """
Artificial Intelligence is transforming industries across the globe.
From healthcare and finance to education and entertainment, AI-powered solutions
are automating tasks, improving accuracy, and enabling faster decision-making.
However, challenges like bias, transparency, and ethical concerns must be addressed
to ensure AI benefits all of humanity.
"""

print(summarize_text(sample_text))
```

```python
import gradio as gr

def ui_summarize(text, min_len, max_len):
    return summarize_text(text, int(min_len), int(max_len))

with gr.Blocks(title="AI Text Summarizer") as demo:
    gr.Markdown("## ✨ AI Text Summarizer\nPaste long text and get a summary.")
    t = gr.Textbox(lines=12, label="Input Text")
    min_len = gr.Slider(20, 160, value=60, step=5, label="Min Length")
    max_len = gr.Slider(60, 240, value=140, step=5, label="Max Length")
    out = gr.Textbox(lines=8, label="Summary")
    btn = gr.Button("Summarize")
    btn.click(ui_summarize, [t, min_len, max_len], out)

demo.launch()
```
